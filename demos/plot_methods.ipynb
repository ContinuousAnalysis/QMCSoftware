# The QMCPY Plot Method

#### This notebook demonstrates the different usages of the plot method. 

##### **Some important parameters of the plot method defined:**
##### *n(int or array): n is the number of samples that will be plotted or a list of samples(used for extensible point sets)*
##### *d_vertical (int or array): d_vertical is the index of points that will be plotted on the vertical axis or a list of indexes of points that will be plotted on multiple vertical axes(for higher dimensional plotting). Default value is 0.*
##### *d_horizontal (int or array): d_horizontal is the index of points that will be plotted as the horizontial axis  or a list of indexes of points that will be plotted on multiple horizontal axes(for higher dimensional plotting). Default value is 1.*


##### 2 dimensional pairs plotting :

```python
n = 2**7 
discrete_distrib = qp.IIDStdUniform(2)
discrete_distrib.plot(n)
```

##### Extensible Point Sets:

```python
n = [2**6, 2**7, 2**8]
discrete_distrib = qp.Halton(2)
discrete_distrib.plot(n)
```

##### High Dimensional Pairs Plotting(All Dimensions):

```python
discrete_distrib = qp.DigitalNetB2(4)
n = 2**7
fig, ax = discrete_distrib.plot(n, d_horizontal = [1,2,3], d_vertical = [0,1,2]) # All Four Dimensions
```

##### High Dimensional Pairs Plotting(Certain Specified Dimensions):

```python
discrete_distrib = qp.DigitalNetB2(4)
n = 2**7
fig, ax = discrete_distrib.plot(n, d_horizontal = [1,2], d_vertical = [0,1]) # Dimensions 1, 2 and 3.
```

##### High Dimensional Pairs Plotting combined with Extensible Point Sets: 

```python
discrete_distrib = qp.Lattice(5)
n = [2**6, 2**7, 2**8]
fig, ax = discrete_distrib.plot(n, d_horizontal = [1,2,3,4], d_vertical = [0,1,2,3])
```
