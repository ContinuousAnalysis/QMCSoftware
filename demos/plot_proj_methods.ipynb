# The QMCPY Plot Projections Method

#### This notebook demonstrates the different usages of the plot_proj  method. 

##### **Some important parameters of the plot_proj method defined:**
##### *n(int or array): n is the number of samples that will be plotted or a list of samples(used for extensible point sets)*
##### *d_vertical (int or array): d_vertical is the index of points that will be plotted on the vertical axis or a list of indexes of points that will be plotted on multiple vertical axes(for higher dimensional plotting). Default value is 1 (dimension 2).*
##### *d_horizontal (int or array): d_horizontal is the index of points that will be plotted as the horizontial axis  or a list of indexes of points that will be plotted on multiple horizontal axes(for higher dimensional plotting). Default value is 0 (dimension 1).*


##### 2 dimensional pairs plotting :

```python
n = 2**7 
iid = qp.IIDStdUniform(2)
iid.plot_proj(n);
```

##### Extensible Point Sets:

```python
n = [2**6, 2**7, 2**8]
Halton = qp.Halton(4)
Halton.plot_proj(n);
fig, ax = Halton.plot_proj(n, d_horizontal = [0,1,2], d_vertical = [1,2,3]) 
```

##### High Dimensional Pairs Plotting(All Dimensions):

```python
net = qp.DigitalNetB2(4)
n = 2**7
fig, ax = net.plot_proj(n, d_horizontal = [0,1,2], d_vertical = [1,2,3]) 
```

##### High Dimensional Pairs Plotting(Certain Specified Dimensions):

```python
discrete_distrib = qp.DigitalNetB2(4)
n = 2**7
fig, ax = discrete_distrib.plot_proj(n, d_horizontal = [0,2], d_vertical = [1,3]) 
```

##### High Dimensional Pairs Plotting combined with Extensible Point Sets: 

```python
lattice = qp.Lattice(5)
n = [2**6, 2**7, 2**8]
fig, ax = lattice.plot_proj(n, d_horizontal = [0,1,2,3], d_vertical = [1,2,3,4])
```
