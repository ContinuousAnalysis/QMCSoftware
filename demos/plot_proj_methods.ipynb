```python
seed = 7
import matplotlib.pyplot as plt
import numpy as np
import qmcpy as qp
iid = qp.IIDStdUniform(2)
Halton = qp.Halton(4)
net = qp.DigitalNetB2(4)
lattice = qp.Lattice(5)
```

# The QMCPY Plot Projections Method For Discrete Distribution

#### This notebook demonstrates the different usages of the plot_proj method. 

##### A discrete distribution object with d dimensions has a maximum of $dP2$ dimensional pairings (for e.g: [2,3] and [3,2] are being considered seperate parings). The plot_proj method plots all or a subset of all the possible dimension pairings using the parameters d_vertical and d_horizontal and can also display extensibility based on the parameter n.

##### **Some important parameters of the plot_proj method defined:**
##### *n(int or array): n is the number of samples that will be plotted or a list of samples(used for extensible point sets)*
##### *d_horizontal (int or array): d_horizontal is a list of dimensions to be plotted on the horizontal axes. For python indices, the possible values are 0 to d-1. For math indices, possible values are from 1 to d. Default value is 0 (1st dimension).*
##### *d_vertical (int or array): d_vertical is a list of dimensions to be plotted on the vertical axes for each corresponding element in d_horizontal.For python indices, the possible values are 1 to d-1. For math indices, possible values are from 2 to d. Default value is 1 (2nd dimension).*
##### *math_ind (boolean) : If  it is set True, the user needs to pass math indices for d_horizontal and d_vertical. Default value is False, indicating the user needs to pass python indices for d_horizontal and d_vertical.*


##### 2 dimensional pairs plotting :

```python
n = 2**7 
fig,ax = iid.plot_proj(n);
```

##### 2 dimensional pairs plotting with Extensible Point Sets:

```python
n = [2**5, 2**6, 2**7]
fig,ax = Halton.plot_proj(n);
```

##### High Dimensional Pairs Plotting(All Dimensions):

```python
n = 2**7
fig, ax = net.plot_proj(n, d_horizontal = [2,3], d_vertical = [2,3,4], math_ind= True);
```

##### High Dimensional Pairs Plotting(Certain Specified Dimensions):

```python
n = 2**7
fig, ax = net.plot_proj(n, d_horizontal = [0,2], d_vertical = [1,3]);
```

##### High Dimensional Pairs Plotting combined with Extensible Point Sets: 

```python
n = [2**6, 2**7, 2**8]
fig, ax = lattice.plot_proj(n, d_horizontal = [0,1,2,3], d_vertical = [1,2,3,4]);
```
